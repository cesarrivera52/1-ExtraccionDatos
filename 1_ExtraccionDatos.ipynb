**Filro básico para variables cuántitativas**

```python
#Filtro por comparación
filtro6=df[df["Periodo"]>= 2006]
filtro6
```

**Filtros basicos con conectores logicos**

```python
#Filtro Y
filtro7= df[(df["Periodo"] > 2015) & (df["Sexo"]== "Hombres")]
filtro7
```

```python
#Filtro O
filtro8= df[(df["Periodo"]>2006) | (df["Sexo"]== "Mujeres")]
filtro8
```

```python
#Filtro not
filtro9=df[~(df["Periodo"] >= 2006) & ~(df["Sexo"]== "Mujeres")]
filtro9
```

```python
#Convertir archivo filtro a csv
filtro9.to_csv("Entregable1.csv")
```
